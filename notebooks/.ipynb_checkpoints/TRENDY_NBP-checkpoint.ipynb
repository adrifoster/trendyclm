```python
import os
import matplotlib.pyplot as plt
import trendy_functions as trendy
```

## TRENDY output

First make the Ascii files requested by TRENDY

This script assumes you have already generated the globally gridded files requested by TRENDY.

```python
# output directory
campaign_dir = "/glade/campaign/cgd/tss/projects/TRENDY2025"
output_dir = os.path.join(campaign_dir, 'outputs')

# experiment suffixes
exps = ['S0', 'S1', 'S2', 'S3']
```

```python
for exp in exps:
    trendy.make_nbp_df(f"{output_dir}/{exp}", exp)
```

## TRENDY Criteria
Check to see if we pass criteria

```python
nbp = trendy.get_all_nbp_df(output_dir, exps)
global_df = nbp[nbp.region == 'Global']
```

```python
trendy.plot_running_means(global_df)
plt.savefig('NBP_running_means.png')
```

```python
trendy.plot_cumulative_nbp(nbp)
plt.savefig('Cumulative_NBP.png')
```

```python
trendy.check_trendy_criteria(global_df, drift_tol=0.05, offset_tol=0.10)
```
